```python
import sympy as sp
from sympy.plotting import plot

x=sp.symbols('x')
```

```python

#put the funtion here
# f1=x
f1=sp.Piecewise(
    (0,(x>-sp.pi)&(x<0)),
    (sp.pi,(x>=0)&(x<sp.pi))
)




s=sp.fourier_series(f1,(x,-sp.pi,sp.pi))
```

```python
y=s[0]
y0=s[0]
```

```python
for i in range(10):
#     si=s[i]
    y0=y0+s[i+1]
#     print(si)
#     print(y0)
y0
```

```python
tuhin=plot(f1,
           s[0],
           s[1]+y,
           s[2]+y,
           s[3]+y,
           s[4]+y,
           s[5]+y,
           s[6]+y,
           s[7]+y,
           s[8]+y,
           s[9]+y,
           s[10]+y,
           y0,
          
           (x,-sp.pi,sp.pi),
           show=False,
           legend=False
           
          )
tuhin[0].line_color = (0.99, 0.55, 0)
tuhin[0].label = 'f1'
tuhin[12].line_color = (0.55, 0.33, 0.66)
tuhin[12].label = 'fourier result'
tuhin[1].line_color = (0.7, 0.7, 0.7)
tuhin[1].label = f'a0/2={s[0]}'
tuhin[2].line_color = (0.7, 0.7, 0.7)
tuhin[2].label = f'a1={s[1]}'
tuhin[3].line_color = (0.7, 0.7, 0.7)
tuhin[3].label = f'b1={s[2]}'
tuhin[4].line_color = (0.7, 0.7, 0.7)
tuhin[4].label = f'a2={s[3]}'
tuhin[5].line_color = (0.7, 0.7, 0.7)
tuhin[5].label = f'b2={s[4]}'
tuhin[6].line_color = (0.7, 0.7, 0.7)
tuhin[6].label = f'a3={s[5]}'
tuhin[7].line_color = (0.7, 0.7, 0.7)
tuhin[7].label = f'b3{s[6]}'
tuhin[8].line_color = (0.7, 0.7, 0.7)
tuhin[8].label = f'a4={s[7]}'
tuhin[9].line_color = (0.7, 0.7, 0.7)
tuhin[9].label = f'b4={s[8]}'
tuhin[10].line_color = (0.7, 0.7, 0.7)
tuhin[10].label = f'a5={s[9]}'
tuhin[11].line_color = (0.7, 0.7, 0.7)
tuhin[11].label = f'b5={s[10]}'

tuhin.show()
```
